<h2 style="text-align:center;">Data scraping: patients' experience and comments</h2>

```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup

header = ['頁數','婦產科醫生','診所地區','專頁','留言日期','求診原因','主題','評價']
info = []

for i in range (1,3):
    url = "https://www.goodoctor.com.hk/comments?search_specialists%5B0%5D=3&page="+str(i)
    quote_page = requests.get(url) 
    soup = BeautifulSoup(quote_page.content,'html.parser') 
    for item in soup.find_all("div", attrs={"class":"doctor-info-item"}):
         page_no = str(i)
         doctor = item.find("div",attrs={"class":"doctor-info-box"}).get_text().strip()
         place = item.find("span",attrs={"class":"area"})
         url = item.find("div",attrs={"class":"doctor-info-box"}).a.get('href')
         date = item.find("span",attrs={"class":"num-family-12px vertical-align-m"})
         reason = item.find("span",attrs={"class":"reason-text"})
         topic = item.find("span",attrs={"class":"title-s"})
         comments = item.find("div", attrs={"class":"comment-content"})
         info.append((page_no, doctor, place, url, date, reason, topic, comments))
df = pd.DataFrame(info,columns=header)
df
```

```python
df = df.drop([1,6,11,15], axis=0)
df.to_csv('info.csv', sep='\t', encoding='utf-8') 
df
```
